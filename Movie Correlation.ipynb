```python
#Import the packages that will be used in this project
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
pd.options.mode.chained_assignment = None

#import the csv 
df = pd.read_csv('/Users/kaizhang/Desktop/Data Portfolio/movies.csv')
df.head()
```

```python
#Check to see if there's any missing values
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing*100)))
#There were missing values, hence delete the rows with null values
df = df.dropna()
print()
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing*100)))
```

```python
#Data Types for the columns
print(df.dtypes)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
df = df.sort_values(by='gross', inplace=False, ascending=False)
```

```python
#To see all the data
#pd.set_option('display.max_rows', None)
```

```python
#Remove any duplicates
df.drop_duplicates()
```

```python
#Hypothesis: There is a high correlation between budget and gross
#Scatter plot with budget vs gross
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget VS. Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
df.head()

```

```python
sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Correlation
#There is a high correlation between budget and gross. 
#Correlation:0.74
```

```python
#Hypothesis: There is a high correlation between company and gross
df.head()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
